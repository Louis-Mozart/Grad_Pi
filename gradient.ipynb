```python
def derivative(f, x, h=1e-5):
    """Compute the numerical derivative of a function f at point x using central difference."""
    return (f(x + h) - f(x - h)) / (2 * h)


def f(x):
    """Example function: f(x) = x^4 - 9x^2 + 15x - 7."""
    return x ** 2 #- 9*x**2 + 15*x - 7

def eta_n(x0,x1):
    eta = abs((x1-x0)*(derivative(f,x0)-derivative(f,x1)))/((derivative(f,x0)-derivative(f,x1))+1e-5)**2
    return eta

x0 = 1.0
for i in range(100):
    x1 = x0 - 0.1 * derivative(f, x0)
    if i % 10 == 0:
        print(f"x{i}:", x0)
    x0 = x1

print(50 * "*")

x0 = 1.0
x1 = x0 - 0.1 * derivative(f, x0)
for i in range(20):
    x2 = x1 - eta_n(x1, x0) * derivative(f, x1)
    # if i % 10 == 0:
    print(f"x{i}:", x1)
    x0 = x1
    x1 = x2

```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4, 4, 400)
y = f(x)
plt.plot(x, y, label='f(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Function Plot')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```
